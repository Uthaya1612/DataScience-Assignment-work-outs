Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.


```python
import numpy as np
mu, sigma = 0.5, 0.2 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
#print(s)
print(abs(mu - np.mean(s)))

abs(sigma - np.std(s, ddof=1))

import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.


The properties of Binominal distribution are
1.There are only two distinct possible outcomes: true/false.
2.here is a fixed number of 'n' times repeated trials in a given experiment.
3.The probability of success or failure remains constant for each attempt/trial.
4.Only the successful attempts are calculated out of 'n' independent trials.
5.Every trial is an independent trial on its own, this means that the outcome of one trial has no effect on the outcome of another trial.

Examples:
1.Tossing a fair coin thrice. Possibilities of getting 2 heads
2.Rolling a dies 5 times. Possibilities of getting exact 6 three times.

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
  
# No of Data points
N = 500
  
# initializing random values
data = np.random.randn(N)
  
# getting data of the histogram
count, bins_count = np.histogram(data, bins=10)
  
# finding the PDF of the histogram using count values
pdf = count / sum(count)
  
# using numpy np.cumsum to calculate the CDF
# We can also find using the PDF values by looping and adding
cdf = np.cumsum(pdf)
  
# plotting PDF and CDF
plt.plot(bins_count[1:], pdf, color="red", label="PDF")
plt.plot(bins_count[1:], cdf, label="CDF")
plt.legend()
```

Q7. How Binomial distribution different from Poisson distribution?


The binomial distribution is one, whose possible number of outcomes are two, i.e. success or failure. On the other hand, there is no limit of possible outcomes in Poisson distribution.

Binomial distribution is one in which the probability of repeated number of trials are studied.
Poisson Distribution gives the count of independent events occur randomly with a given period of time.
