Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

In ArithmeticError all errors related arithmetic operations will come. Main classifications are
1. ZeroDivisionError - Numerical divided by Zero gives Zero Division Error.
2. ValueError - Raised when a built-in function receives an argument that has the right type but an inappropriate value
3. TypeError -Passing arguments of the wrong type should result in a TypeError
4. OverflowError - When numerical index value goes beyond 34 - overflow error occurs

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print (j)
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

Q4.Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary respectively.

```python
# key error occurs when the invoked key not present in the dictionary list.
try :
    d = {1: [3,4,5,6] , "key" :"sudh"}
    d["key10"]
except KeyError as e : 
    print(e)
```

```python
# Index error occurs when the invoked index is out of range in list.
try :
    d = [1,2,3,4]
    d[10]
except IndexError as e : 
    print(e)
```
